```python
from glob import glob
stock_files=glob('val(word).csv')
stock_files
```

```python
dataset=pd.concat((pd.read_csv(file).assign(filename = file)
          for file in stock_files), ignore_index=True)
dataset
```

```python
dataset
```

```python
f1scores =[]
for num in range(0,101):
  tp=0
  tn=0
  fp=0
  fn=0
  for i in range(dataset.shape[0]):
    if dataset['round(Trigram)'][i]>num:
      if dataset['label'][i] == 1:
        tp+=1
      else:
        fp+=1
    else:
      if dataset['label'][i]== 0:
        tn+=1
      else:
        fn+=1
  f1score = tp/(tp+0.5*(fp+fn))
  f1scores.append(f1score) 
```

```python
mx=max(f1scores)
print(max(f1scores))
loc=f1scores.index(mx)
print(loc)
```

for n-grams

```python
pred = []
label = []
for i in range(dataset.shape[0]):
    label.append(dataset['label'][i])
    if dataset['round(Trigram)'][i]>loc:
      pred.append(1)
    else:
      pred.append(0)

dataset['pred(Trigram)']=pred
```

for combind ones

```python
pred = []
label = []
for i in range(dataset.shape[0]):
    label.append(dataset['label'][i])
    pred.append(dataset['pred(Unigram+Bigram+Trigram)'][i])
```

```python
import sklearn
```

```python
sklearn.metrics.f1_score(label,pred)
```

```python
print(accuracy_score(label,pred))
print(precision_score(label,pred))
print(recall_score(label,pred))
```

```python
dataset.to_csv('val(word).csv')
```
